```python
import pandas as pd

df = pd.read_csv('results/supplier_time.csv')
df
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [9, 6]
plt.rcParams["figure.autolayout"] = True

axes = df.plot()
axes.get_legend().remove()
plt.ylabel('Time to Execute (ms)', fontsize = 11)
plt.xlabel('Transactions', fontsize = 11)

plt.savefig('Supplier Transaction Times.pdf')
plt.show()

```

```python
df = pd.read_csv('results/client_time.csv')
df
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [9, 3]
plt.rcParams["figure.autolayout"] = True

axes = df.plot()
axes.get_legend().remove()
axes.set_ylim([300, 650])
plt.ylabel('Time to Execute (ms)', fontsize = 11)
plt.xlabel('Hour of Day', fontsize = 11)

plt.savefig('Client Transaction Times.pdf')
plt.show()

```
